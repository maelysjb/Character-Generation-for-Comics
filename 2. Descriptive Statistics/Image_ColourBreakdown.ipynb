# Colour 

Every digital image is made up of pixels and each of those pixel consists of three channels/layers. The values of intensity of the channels vary from 0-255 and each number is an 8-bit number. When the channels are muted they are closer to that 0 number and they show as white, then when the values are closer to full intensity or 255, they appear as black. 

The three main colour channels: 
- Red (R)
- Green (G)
- Blue (B)

### Imports 

```python
# !pip install imageio
import imageio
import matplotlib.pyplot as plt
%matplotlib inline
pic = imageio.imread('1. Data/sks-03.jpeg')
plt.figure(figsize = (5,5))
plt.imshow(pic)
```

### Basic Properties of Image

```python
print('Type of the image : ' , type(pic)) 
print('Shape of the image : {}'.format(pic.shape)) 
print('Image Hight {}'.format(pic.shape[0])) 
print('Image Width {}'.format(pic.shape[1])) 
print('Dimension of Image {}'.format(pic.ndim))
```

```python
print('Image size {}'.format(pic.size)) 
print('Maximum RGB value in this image {}'.format(pic.max())) 
print('Minimum RGB value in this image {}'.format(pic.min()))
```

```python
# A specific pixel located at Row : 200 ; Column : 100  
# Each channel's value of it, gradually R , G , B  
print('Value of only R channel {}'.format(pic[ 100, 50, 0])) 
print('Value of only G channel {}'.format(pic[ 100, 50, 1])) 
print('Value of only B channel {}'.format(pic[ 100, 50, 2]))
```

### Each channel in the whole image
Red (R) channel

```python
plt.title('Red channel') 
plt.ylabel('Height {}'.format(pic.shape[0])) 
plt.xlabel('Width {}'.format(pic.shape[1])) 
plt.imshow(pic[ : , : , 0], cmap='gray')
plt.show()
```

Green (G) channel 

```python
plt.title('Green channel')
plt.ylabel('Height {}'.format(pic.shape[0])) 
plt.xlabel('Width {}'.format(pic.shape[1])) 
plt.imshow(pic[ : , : , 1], cmap='gray') 
plt.show()
```

Blue (B) channel

```python
plt.title('Blue channel') 
plt.ylabel('Height {}'.format(pic.shape[0])) 
plt.xlabel('Width {}'.format(pic.shape[1])) 
plt.imshow(pic[ : , : , 2], cmap='gray') 
plt.show()
```

### Applying intensity of each channel 

```python
# Red (R) channel 
pic[300:600 , : , 0] = 255 # full intensity to those pixel's R channel 
plt.figure( figsize = (5,5)) 
plt.imshow(pic) 
plt.show()
```

```python
# Green (G) channel
pic[800:1100 , : , 1] = 255 # full intensity to those pixel's G channel 
plt.figure( figsize = (5,5)) 
plt.imshow(pic) 
plt.show()
```

```python
# Blue (B) channel 
pic[1200:1500 , : , 2] = 255 # full intensity to those pixel's B channel 
plt.figure( figsize = (5,5)) 
plt.imshow(pic) 
plt.show()
```

### Splitting Layers 

```python
import numpy as np 
pic = imageio.imread('1. Data/sks-03.jpeg') 
fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15,5))  
for c, ax in zip(range(3), ax):     
     # create zero matrix        
     split_img = np.zeros(pic.shape, dtype="uint8") 
     # 'dtype' by default: 'numpy.float64'  # assing each channel      
     split_img[ :, :, c] = pic[ :, :, c] # display each channel     
     ax.imshow(split_img)
```
