```python
import os
import json
from PIL import Image
import json
from datasets import load_dataset
import pandas as pd 

# Modify your training script to use the resized images and JSON file
MODEL_NAME = "CompVis/stable-diffusion-v1-4"
OUTPUT_DIR = "/kaggle/working/outputs"
HUB_MODEL_ID = "nataliabeltran/LoRA1000"
# DATASET_DIR = "/kaggle/input/practicetext2image/"

# dataset_name = "linoyts/Tuxemon"
dataset_name = "nataliabeltran/UnicornGirl"
```

```python
from accelerate.utils import write_basic_config

write_basic_config()
```

```python
# TRAINING 
!accelerate launch  /kaggle/working/diffusers/examples/text_to_image/train_text_to_image_lora.py \
  --mixed_precision="fp16" \
  --pretrained_model_name_or_path=$MODEL_NAME \
  --dataset_name=$dataset_name \
  --caption_column='caption' \
  --image_column='image' \
  --dataloader_num_workers=8 \
  --resolution=512 \
  --center_crop \
  --random_flip \
  --train_batch_size=6 \
  --gradient_accumulation_steps=10 \
  --max_train_steps=1000 \
  --learning_rate=1e-05 \
  --max_grad_norm=1 \
  --lr_scheduler="cosine" \
  --lr_warmup_steps=0 \
  --output_dir=$OUTPUT_DIR \
  --push_to_hub \
  --hub_model_id=$HUB_MODEL_ID \
  --checkpointing_steps=900 \
  --validation_prompt="An image of UnicornGirl in unicorn onesie" 
```

```python
from diffusers import AutoPipelineForText2Image
import torch

pipeline = AutoPipelineForText2Image.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16).to("cuda")
pipeline.load_lora_weights("/kaggle/working/outputs/", weight_name="pytorch_lora_weights.safetensors")
#image = pipeline("An image of UnicornGirl in unicorn onesie running").images[0]
```
